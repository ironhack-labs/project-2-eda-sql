```python
frequency_table_dist = cat.market_segment.value_counts() # Hotel distribuition
proportion_table_dist = cat.market_segment.value_counts(normalize=True).round(2)
```

```python
df_dist = (pd.concat([frequency_table_dist, proportion_table_dist], axis=1)*100)
df_dist.columns = ['absolute_frequency', 'relative_frequency']
df_dist['relative_frequency'] = df_dist['relative_frequency'].astype(str) + '%'
df_dist
```

```python
pd.crosstab(index=cat.reservation_status,
           columns='count')
```

```python
crosstab_result_resv = pd.crosstab(index=cat.reservation_status, columns=cat.hotel)
crosstab_result_resv 
```

```python
frequency_table_resv = cat.reservation_status.value_counts() # counting the different Reservation Status
proportion_table_resv = cat.reservation_status.value_counts(normalize=True).round(2) # converting the countings in the proportional part
```

```python
df_resv = pd.concat([frequency_table_resv, proportion_table_resv], axis=1)*100 # convert to percentage
df_resv.columns = ['absolute_frequency', 'relative_frequency']
df_resv['relative_frequency'] = df_resv['relative_frequency'].apply(lambda x: f"{x}%")  #appling % 
df_resv
```

```python
df_resv_adr = df.groupby(["hotel", "reservation_status"])["adr"].agg(["median", "mean"])  # AvDailyRate - count, mean, median
round(df_resv_adr)
```

```python
df_resv_adr_r = df_resv_adr.round()
df_resv_adr_r['median'] = df_resv_adr_r['median'].apply(lambda x: f'{x:.1f}€')
df_resv_adr_r['mean'] = df_resv_adr_r['mean'].apply(lambda x: f'{x:.1f}€')

df_resv_adr_r
```

```python
df_resv_rn = df.groupby(["hotel", "reservation_status"])["rn_sum"].agg(["count", "median", "mean"])  # RoomNights - count, mean(avStayby Reservation, median
round(df_resv_rn)
```

```python
df_resv_dist = df.groupby(["hotel", "distribution_channel"])["rn_sum",].agg(["count", "median", "mean", "std"])  # RoomNights - count, mean, median
df_resv_dist_r = df_resv_dist.round(1)
df_resv_dist_r
```

```python
df_resv_meal = df.groupby(["hotel", "meal"])["rn_sum"].agg(["count", "median", "mean", "std"])  # RoomNights - count, mean, median
df_resv_meal_r = df_resv_meal.round(1)
df_resv_meal_r
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['hotel'], df['is_canceled'])  # perform a Chi square, howhever i dont know so much about it

chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {round(chi2),0}")
print(f"P-value: {p}")
```
